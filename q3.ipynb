```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("data.csv", sep=";")
```

```python
clean_df = df[["Meanpart", "REP_DR2"]].dropna()

elec_part = clean_df["Meanpart"]
wom_rep = clean_df["REP_DR2"]

m, b = np.polyfit(elec_part, wom_rep, 1)

plt.plot(elec_part, m * elec_part + b, color="red")

plt.scatter(elec_part, wom_rep)
```
