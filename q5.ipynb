## Does a higher number of effective parties lead to government instability?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("data.csv", sep=";")

clean_df = df[["Enep", "Govstab"]].dropna()
x = clean_df["Enep"]
y = clean_df["Govstab"]

m, b = np.polyfit(x, y,1)

x_line = np.linspace(x.min(), x.max(), 2)
y_line = m*x_line + b

plt.plot(x_line, y_line, color="red")
plt.scatter(x,y)
```

```python
most_parties = df[["Country", "Year", "Enep", "Govstab", "Cabchange"]].nlargest(10, "Enep")

print(most_parties)
```

## Does military involvement in politics decrease physical integrity rights?

```python
clean_df = df[["IL_PHIN", "Mip"]].dropna()

x = 8 - clean_df["Mip"]
y = clean_df["IL_PHIN"]


m, b = np.polyfit(x,y,1)

plt.plot(x, m*x +b, color="red")


plt.scatter(x,y)
plt.xlabel("Military intervention")
plt.ylabel("Physical integrity rights")
```
